```python
#Medida Jaccard que mide la proporción de elementos de un conjunto en otro
def jaccard(s1: set, s2: set):
    return len(s1 & s2) / len(s1 | s2)  
```

```python
def coinsidencia(conjuntoA, conjuntoB):
    matches = []
    for palabra in conjuntoA:
        if palabra in conjuntoB:
            matches.append(palabra)
    conteo = len(matches)
    return matches, conteo
```

### Resultados Palabras Asela

```python
jaccgpt = jaccard(set(df['Unnamed: 0']), set(palabrasProtoAsela['Palabra']), )

print('Jaccard GPT:', jaccgpt)
```

```python
jaccgptpsiq = jaccard(set(df['Unnamed: 0']), set(palabrasPsiq['word']))

print('Jaccard GPT:', jaccgptpsiq)
```

```python
matches, conteos = coinsidencia(list(df['Unnamed: 0']), list(palabrasPsiq['word']))

print(matches)
print(conteos)
```
